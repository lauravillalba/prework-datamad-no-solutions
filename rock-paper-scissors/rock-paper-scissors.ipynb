# Rock Paper Scissors

Let's play the famous game against our computer.

https://en.wikipedia.org/wiki/Rock%E2%80%93paper%E2%80%93scissors

The use of functions is recommended

## Goals

2. Data capture by console

5. Definition of functions. Modular programming
6. Logical operators.
7. Print
8. Import modules

```python
total=5
wins_to_win=2
games=0
wins_gamer=0
wins_machine=0
tie=0

print('RULES: \n- Win the best of', total, "games! \n- You have to win 2 games minimum.\n\n>>>Let's play!!")
while games<total:
  gamer=input("\nrock, paper o scissors -> take your choice: ")

  if (gamer== 'rock') or (gamer=='paper') or (gamer=='scissors'):
      import random
      random_machine = ['rock', 'paper', 'scissors']
      machine = random.choice(random_machine)
      print('machine choice:', machine)

      games+=1
     
      if gamer==machine:
        tie+=1
        print ('TIE!--> 0')
      elif (gamer=='rock')and(machine=='scissors'):
        wins_gamer+=1
        print ('YOU WIN!--> 2')
      elif (gamer=='paper')and (machine=='rock'):
        wins_gamer+=1
        print ('YOU WIN!--> 2')
      elif (gamer=='scissors')and(machine=='paper'):
        wins_gamer+=1
        print ('YOU WIN!--> 2')
      else:
        wins_machine+=1
        print ('MACHINE WINS! --> 1')
    
  else:
      print ('Por favor, introduce una de las tres opciones mostradas:')

if (wins_gamer<wins_to_win) and (wins_machine<wins_to_win):
  print ('\n\n <<<<No winner, play again!>>>>')
elif wins_gamer > wins_machine:
  print('\n\n <<<<You WIN the GAME!>>>>')
elif wins_gamer<wins_machine:
  print ('\n\n<<<<Machine WINS the GAME!>>>>')
```
